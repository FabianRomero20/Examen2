```python
# Importamos nuestra Tabla de datos

import pandas as pd
import os

dataBol=pd.read_excel(os.path.join('data','dataBolivia_indicadores.xlsx'))
dataBol.head()
```

```python
# Ahora, importamos nuestra Mapa

os.environ['USE_PYGEOS'] = '0'

import geopandas as gpd

dataBolmap=gpd.read_file(os.path.join('maps','Provincias.zip'))

dataBolmap.head()
```

```python
# Realizamos el Preprocessing
# Igualamos las mayúsculas y espacios en blanco de las columnas (departamentos y provincias) para un mejor trabajo
# Para ambas variables

capitalizeColumns=lambda x: x.str.upper().str.strip()
dataBol[['DEPARTAMENTO','PROVINCIA']]=dataBol[['DEPARTAMENTO','PROVINCIA']].apply(capitalizeColumns)
dataBolmap[['DEPARTAMEN','PROVINCIA']]=dataBolmap[['DEPARTAMEN','PROVINCIA']].apply(capitalizeColumns)
```

```python
# Quitamos algunos símbolos comúnes en los nombres de nuestros departamentos y provincias
# Para ambas variables

import unidecode

byePunctuation=lambda x: unidecode.unidecode(x)
dataBol[['DEPARTAMENTO','PROVINCIA']]=dataBol[['DEPARTAMENTO','PROVINCIA']].applymap(byePunctuation)
dataBolmap[['DEPARTAMEN','PROVINCIA']]=dataBolmap[['DEPARTAMEN','PROVINCIA']].applymap(byePunctuation)
```

```python
# Como trabajaremos con las Provincias de Bolivia
# Observamos si obtenemos la misma cantidad de Provincias en ambas variables

len(dataBol.PROVINCIA),len(dataBolmap.PROVINCIA)
```

```python
# Observamos la cantidad de nombres de Provincias repetidos o duplicados
# Para ambas variables

dataBol.PROVINCIA.duplicated().sum(),dataBolmap.PROVINCIA.duplicated().sum()
```

```python
# Debido a la presencia de nombre duplicados
# Optamos por crear una nueva columna para diferenciarlos mejor
# Realizamos la concatenating

dataBol['provDist']=["+".join(pd) for pd in zip (dataBol.DEPARTAMENTO,dataBol.PROVINCIA)]
dataBolmap['provDist']=["+".join(pd) for pd in zip (dataBolmap.DEPARTAMEN,dataBolmap.PROVINCIA)]
```

```python
# Observamos la nueva columna

dataBol['provDist'].head()
```

```python
# Reemplazamos guiones y espacios múltiples por un espacio simple
# Para ambas variables

dataBol.provDist=dataBol.provDist.str.replace("\-|\_|\s+"," ",regex=True)
dataBolmap.provDist=dataBolmap.provDist.str.replace("\-|\_|\s+"," ",regex=True)
```

```python
# Realizamos un dendograma con nuestras variables 

from scipy.cluster import hierarchy as hc

# El método de clasificación que se usará será el Fisherjenks
# Entonces, para obtener el algoritmo de agrupamiento que se usará
# Hacemos lo siguiente

# Obtenemos los valores de los datos de interés
datos = dataBolmap[['ViviendasNOPropiasPct', 'ViviendasConAguaPct', 'ViviendasConEnergíaPct']]

# Esto generará los límites de los intervalos óptimos para cada columna:
k = 5  # Número deseado de grupos

limites_columna1 = mapclassify.FisherJenks(datos['ViviendasNOPropiasPct'], k=k).bins
limites_columna2 = mapclassify.FisherJenks(datos['ViviendasConAguaPct'], k=k).bins
limites_columna3 = mapclassify.FisherJenks(datos['ViviendasConEnergíaPct'], k=k).bins


# Combinamos los límites de los intervalos de todas las columnas en una sola lista
limites_totales = np.concatenate([limites_columna1, limites_columna2, limites_columna3])


# Utilizamos los límites totales para convertir cada columna en grupos discretos utilizando 
grupos_columna1 = np.digitize(datos['ViviendasNOPropiasPct'], limites_columna1)
grupos_columna2 = np.digitize(datos['ViviendasConAguaPct'], limites_columna2)
grupos_columna3 = np.digitize(datos['ViviendasConEnergíaPct'], limites_columna3)

# Combinamos los grupos de todas las columnas en un solo arreglo multidimensional
grupos_totales = np.column_stack([grupos_columna1, grupos_columna2, grupos_columna3])

# Finalmente, pasamos a realizar nuestro dendograma
Z = hc.linkage(grupos_totales, method='ward')
# Calculamos el dendrograma completo
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('cases')
plt.ylabel('distance')
hc.dendrogram(
    Z,
    leaf_rotation=90.,  # giramos las etiquetas del eje x
    leaf_font_size=1,  # tamaño de fuente para las etiquetas del eje x
)
plt.show()
```

```python
# Decidimos cuántos grupos se pueden crear usando esos datos

from sklearn.cluster import AgglomerativeClustering as agnes

# Establecemos la semilla para la reproducibilidad
np.random.seed(12345)

# Inicializamos el algoritmo, solicitando 3 clústeres
model = agnes(linkage="ward", n_clusters=3).fit(dataBolmap[selected_variables_new])

# Asignamos etiquetas a la tabla de datos principal
dataBolmap["hc_3"] = model.labels_
```

```python
# Necesario para la correlación espacial

w_knn.transform = 'R'
```

```python
# La correlación espacial se mide mediante la estadística de Moran

from esda.moran import Moran

# Variables a usar:
# ViviendasPropiasPct ViviendasConAguaPct ViviendasConEnergíaPct

# Primero: ViviendasPropiasPct
moranViviendasPropiasPct = Moran(dataBolmap['ViviendasPropiasPct'], w_knn)
moranViviendasPropiasPct.I,moranViviendasPropiasPct.p_sim
```

```python
from splot.esda import moran_scatterplot

# Veamos el gráfico de dispersión espacial

fig, ax = moran_scatterplot(moranViviendasPropiasPct, aspect_equal=True)
ax.set_xlabel('ViviendasPropiasPct_std')
ax.set_ylabel('SpatialLag_ViviendasPropiasPct_std')
plt.show()
```

```python
# Segundo: ViviendasConAguaPct

moranViviendasConAguaPct = Moran(dataBolmap['ViviendasConAguaPct'], w_knn)
moranViviendasConAguaPct.I,moranViviendasConAguaPct.p_sim
```

```python
# Veamos el gráfico de dispersión espacial

fig, ax = moran_scatterplot(moranViviendasConAguaPct, aspect_equal=True)
ax.set_xlabel('ViviendasConAguaPct_std')
ax.set_ylabel('SpatialLag_ViviendasConAguaPct_std')
plt.show()
```

```python
# Tercero: ViviendasConEnergíaPct

moranViviendasConEnergíaPct = Moran(dataBolmap['ViviendasConEnergíaPct'], w_knn)
moranViviendasConEnergíaPct.I,moranViviendasConEnergíaPct.p_sim
```

```python
# Veamos el gráfico de dispersión espacial

fig, ax = moran_scatterplot(moranViviendasConEnergíaPct, aspect_equal=True)
ax.set_xlabel('ViviendasConEnergíaPct_std')
ax.set_ylabel('SpatialLag_ViviendasConEnergíaPct_std')
plt.show()
```

```python
# El diagrama de dispersión con información local

from esda.moran import Moran_Local

# Calculamos el Moran_Local y hacemos el plot

lisaViviendasConAguaPct = Moran_Local(y=dataBolmap['ViviendasConAguaPct'], w=w_knn,seed=2022)
fig, ax = moran_scatterplot(lisaViviendasConAguaPct,p=0.05)
ax.set_xlabel('ViviendasConAguaPct_std')
ax.set_ylabel('SpatialLag_ViviendasConAguaPct_std')
plt.show()
```

```python
# El mapa con los spots y los outliers

from splot.esda import lisa_cluster
f, ax = plt.subplots(1, figsize=(12, 12))
plt.title('Spots and Outliers')
fig = lisa_cluster(lisaViviendasConAguaPct, 
                   dataBolmap,ax=ax,
                   legend_kwds={'loc': 'center left', 
                                'bbox_to_anchor': (0.7, 0.6)})
plt.show()
```

```python
# Cuadrante
lisaViviendasConAguaPct.q
```

```python
# Significancia
lisaViviendasConAguaPct.p_sim
```

```python
# Cuadrante: 1 HH,  2 LH,  3 LL,  4 HL
pd.Series(lisaViviendasConAguaPct.q).value_counts()
```

```python
# Debemos agregar si la correlación espacial local es significativa

dataBolmap['ViviendasConAguaPct_quadrant']=[l if p <0.05 else 0 for l,p in zip(lisaViviendasConAguaPct.q,lisaViviendasConAguaPct.p_sim)  ]
dataBolmap['ViviendasConAguaPct_quadrant'].value_counts()
```

```python
# Ahora recodificamos

labels = [ '0 no_sig', '1 hotSpot', '2 coldOutlier', '3 coldSpot', '4 hotOutlier']

dataBolmap['ViviendasConAguaPct_quadrant_names']=[labels[i] for i in dataBolmap['ViviendasConAguaPct_quadrant']]

dataBolmap['ViviendasConAguaPct_quadrant_names'].value_counts()
```

```python
# Realizamos el plot

from matplotlib import colors
myColMap = colors.ListedColormap([ 'ghostwhite', 'red', 'green', 'black','orange'])

# Configurar figura y ax
f, ax = plt.subplots(1, figsize=(12,12))

# Trazar valores únicos de coropletas incluyendo
# una leyenda y sin límites
plt.title('Spots and Outliers')

dataBolmap.plot(column='ViviendasConAguaPct_quadrant_names', 
                categorical=True,
                cmap=myColMap,
                linewidth=0.1, 
                edgecolor='black',
                legend=True,
                legend_kwds={'loc': 'center left', 
                             'bbox_to_anchor': (0.7, 0.6)},
                ax=ax)
# Eliminamos eje
ax.set_axis_off()
# Mostramos el mapa
plt.show()
```
